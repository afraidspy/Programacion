## Listas
es una colección ordenada de elementos que permite guardar 
varios valores dentro de una sola variable

```python
nombres = ["Ana", "Luis", "Marta"]
print("Lista: ", nombres)
```

```python
## Acceso a los elementos de una lista
print("Primer elemento ", nombres[0])
```

```python
# -3     -2        -1
["Ana", "Luis", "Marta"]
print("Ultimo elemento ", nombres[-1])
print("Ultimo elemento ", nombres[2])
```

```python
print("Valor de enmedio: ", nombres[-2])
print("Valor de enmedio: ", nombres[1])
```

```python
nombres[0] = "Julieta"  #Una lista es mutable
print(nombres)
```

```python
## Creación de listas
numeros = [10, 20, 30, 40, 50]
lista_combinada =  ["Ana", 25, True, 1.75]
print("Lista combinada: ", lista_combinada)
```

```python
#Conocer el tamaño de una lista
tamanio =  len(numeros)
print("Elementos en la lista numeros: ", tamanio)
```

```python
#Recorrer una lista con un for

for elemento in lista_combinada:
    print(elemento)
```

```python
for indice in range(len(lista_combinada)):
    print(indice, lista_combinada[indice])
```

```python
## Operaciones básica
grupo_1 = ["Diana", "Claudia"]
grupo_2 = ["Rodrigo", "Ricardo"]
todos = grupo_1 + grupo_2
print(todos)
```

```python
repetidos = [1, 2, 3] * 3
print(repetidos)
```

```python
## Slicing
valores = [10, 20, 30, 40, 50, 60]
print("Sublista [1:4]", valores[1:4]) #posiciones= 1,2,3
print("Sublista [0:3]", valores[0:3])
print("Sublista [0:3]", valores[:3])
print("Sublista [3:]", valores[3: ])
```

```python
## Métodos dentro de una lista
#Append
nombres = ["Diana", "Claudia"]
nombres.append("Laura")

print(nombres)
```

```python
## Agregar dentro de una lista en una posicion específica
nombres.insert(1, "Julia")
print(nombres)
```

```python
##Extend
nombres.extend(["Elena","José","Rosa"])
print(nombres)
```

```python
#Remove --> Permite eliminar un elemento de la lista
nombres.remove("Rosa")
print(nombres)
```

```python
#nombres.remove("Jessica")
#print(nombres)
```

```python
#Método pop()
eliminado = nombres.pop()
print("Nombre eliminado: ", eliminado)
print("Lista despues de la eliminación: ", nombres)
```

```python
#Método
nombres.append("Karina")
print(nombres)
```

```python
#Método index
print("Posicion del elemento: ", nombres.index("Julia"))
```

```python
#print("Posicion del elemento: ", nombres.index("Brenda"))
```

```python
numeros_desordenados = [5, 1, 4, 2, 3]
numeros_desordenados.sort()
print(numeros_desordenados)
```

```python
#Reversa de una lista
numeros_desordenados.reverse()
print(numeros_desordenados)
```
